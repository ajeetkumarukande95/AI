```python
# Basic String Operations
# 1.Concatenation
# Combining two or more strings using the + operator.
greeting = "Hello"
name = "Alice"
message = greeting + ", " + name + "!"
print(message)  # Output: Hello, Alice!
```

```python
# Repetition
# Repeating a string using the * operator.
line = "-" * 20
print(line)  # Output: --------------------
```

```python
# Length of a String
# Finding the number of characters in a string using the len() function.

text = "Python"
print(len(text))  # Output: 6
```

```python
# Indexing
# Accessing individual characters in a string using their index (starts from 0).

text = "Python"
print(text[0])  # Output: P
print(text[-1])  # Output: n (negative index starts from the end)
```

```python
# Slicing
# Extracting a substring using slicing.

text = "Python Programming"
print(text[0:6])  # Output: Python
print(text[7:])   # Output: Programming
```

```python
# f-Strings (Python 3.6+)
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old.")
# Output: My name is Alice and I am 25 years old.
```

```python
# format() Method
name = "Alice"
age = 25
print("My name is {} and I am {} years old.".format(name, age))
# Output: My name is Alice and I am 25 years old.
```

```python
# % Operator (Older Style)
name = "Alice"
age = 25
print("My name is %s and I am %d years old." % (name, age))
# Output: My name is Alice and I am 25 years old.
```
